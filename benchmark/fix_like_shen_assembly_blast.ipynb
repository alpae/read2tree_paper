```python
for f in glob.glob('/Users/daviddylus/Projects/r2t/benchmark/blast/assembly_like_shen/*_ogs.txt'):
    fopen = open(f, "r")
    lines = fopen.readlines()
    fopen.close()

    fopen = open(f.replace('_ogs.txt', '_ogs_like_shen.txt'), "r")
    lines_like_shen = fopen.readlines()
    fopen.close()

    x = [l.split(' ')[0].rstrip('\n') for l in lines]
    y = [l.split(' ')[0].rstrip('\n').replace('final', 'fa') for l in lines_like_shen]

    x_og = [l.split(':')[0].replace('like_shen/','') for l in x]
    y_og = [l.split(':')[0].replace('like_shen/','') for l in y]
    
    print(f)
    print(len(set(y_og).intersection(set(x_og))))
    print(len(set(y_og)))
    
    out_lines = []
    fo = open(f.replace('_ogs.txt', '_ogs_like_shen.txt.2'), "w")
    for y1,y2 in zip(y, y_og):
        for x1,x2 in zip(x, x_og):
            if y2 in x2:
                fo.write(x1+'\n')
                out_lines.append(x1)

    fo.close()
```
